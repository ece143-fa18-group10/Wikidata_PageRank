```python
def visualization():
    import pandas as pd
    import numpy as np
    pd.read_csv('link_list.csv')
    from pandas import DataFrame as df
    data = pd.read_csv('link_list.csv')
    data.columns = ['Unnamed: 0', 'source', 'target']
    print(type(data))
    data.target[0]
    ## read PR_result
    with open('PageRank_result.txt', 'r') as pr_result:
        lines = pr_result.readlines()
        size = len(lines)
        page_rank_res = np.empty(size)
        for i in range(size):
            page_rank_res[i] = lines[i]
    print(page_rank_res)
    print(len(data.source))
    index = range(len(page_rank_res))
    #print(index)
    #print(index[1])
    page_rank = dict(zip(index,page_rank_res))
    #print(page_rank)
    page_rank_col = np.empty(len(data.source))
    for i in range(len(data.source)):
        page_rank_col[i] = page_rank[data.source[i]]
    print(page_rank_col)
    print(page_rank_col.shape)
    data['scores'] = page_rank_col
    data.drop('Unnamed: 0', axis = 1)
    with open('result.csv', 'w') as output:
        for i in range(len(data)):
            output.write(str(data.source[i]) + ',' + str(data.target[i]) + ',' + str(data.scores[i]) + '\n')
    return data
```

```python
data = visualization()
print(data)
```
